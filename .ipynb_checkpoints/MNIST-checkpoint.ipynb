```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
import random
```

```python
np.random.seed(0)
```

```python
# _train -> training data sets
# _test -> testing data sets
# mnist.load_data -> produces 60000 training images and 10000 test images that are 28 * 28 sized 
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# assert() -> takes a single argument. If the condition is not met, it'll raise a flag. Used for debugging
## .shape() -> returns (no._of_images, width, height)
assert(X_train.shape[0] == y_train.shape[0]), "The number of the training images is not equal to the number of labels"
assert(X_test.shape[0] == y_test.shape[0]), "The number of the testing images is not equal to the number of labels"
assert(X_train.shape[1:] == (28, 28)), "The dimensions of the training images are not 28x28"
assert(X_test.shape[1:] == (28, 28)), "The dimensions of the testing images are not 28x28"
```

```python
# num_samples -> used to record images in each of our 10 category ( 0 -9 )
# cols -> grid of 10*5 of random images (handwritten digits)
# plt.subplots -> display multiple plots on the same figure and returns a tuple of instance of our figure
## and plot axes
# fig.tight_layout -> avoids overlap between the images
# cmap -> color map; used in here to get a gray colored image

num_of_samples = []
cols = 5
num_classes = 10
fig, axs = plt.subplots(nrows = num_classes, ncols = cols, figsize = (10, 10) )
fig.tight_layout()

for i in range(cols) :
    for j in range(num_classes) :
        # x_selected is populated by X_train training images that corresponds to the value of j, by logically checking its equivalence with the labels
        x_selected = X_train[y_train == j]
        # x_selected will contain an array of images corresponding to a class that equals the value of j and we show a random image in our grid
        axs[j][i].imshow(x_selected[random.randint(0, len(x_selected -1)), :, :], cmap=plt.get_cmap("gray"))
        #  axs[j][i].axis("off")
        if i == 2 :
            num_of_samples.append(len(x_selected))
```

# We plot the number of images in each category vs class

```python
print(num_of_samples)
plt.figure(figsize =(12, 4))
plt.bar(range(0, num_classes), num_of_samples)
plt.title("Distribution of the training dataset")
plt.xlabel("Class number")
plt.ylabel("Number of images")
```

# To perform one hot encoding

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

# Normalize the data values in order to have a range of (0,1), hence 255

```python
# Since the data is moved between layers and also fed to activation function, normalizing them decreases
# the overall variance of data

X_train = X_train/255
X_test = X_test/255
```

# To perform matrix mul, we arrange the matrices accordingly

```python
num_pixels = 784
X_train = X_train.reshape(X_train.shape[0], num_pixels)
X_test = X_test.reshape(X_test.shape[0], num_pixels)
```

# Training our neural network

```python
def create_model() :
    model = Sequential()
    model.add(Dense(10, input_dim=num_pixels, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))                                #o/p layer
    model.compile(Adam(lr=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = create_model()
print(model.summary())
```

```python
# validation_split -> We had to split our data set into training and validation sets ( here 10% )
history = model.fit(X_train, y_train, validation_split=0.1, epochs=10, batch_size = 200, verbose = 1, shuffle = 1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.title('loss')
plt.xlabel('epoch')
            
```

## The more the epochs, at some point the training error will be less than the validation error implying overfitting

```python
# We set verbose = 0 to not display the progress bar information

score = model.evaluate(X_test, y_test, verbose=0)
print('Test score', score[0])
print('Test accuracy', score[1])
```

```python
# We set 'stream' as True so that the 'request' doesn't get downloaded into the memory

import requests
from PIL import Image
url = 'https://www.researchgate.net/profile/Jose_Sempere/publication/221258631/figure/fig1/AS:305526891139075@1449854695342/Handwritten-digit-2.png'
response = requests.get(url, stream = True)
img = Image.open(response.raw)
```

```python
# img_array -> .asarray() takes an argument which it converts to an array
# resized -> We use OpenCV library to resize the image to 28 by 28 since our network was trained to latter size
# gray_scale -> We convert the color image to gray image so that it is 2D
# image -> Our network was trained with images with black backround and white foreground images, hence
## we convert the current image to the ones before by 'inverting' the pixels or by subtracting '255'.
import cv2

img_array = np.asarray(img)
resized = cv2.resize(img_array, (28,28))
gray_scale = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
image = cv2.bitwise_not(gray_scale)
# plt.imshow(image, cmap=plt.get_cmap("gray"))
```

```python
# We normalize the image

image = image/255
image = image.reshape(1, 784)
```

## We use .predict_classes() for multiclass network

```python
 prediction = model.predict_classes(image)
print("predicted digit:", str(prediction))
```
